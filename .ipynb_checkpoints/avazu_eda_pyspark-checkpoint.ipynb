```python
from pyspark.sql import SparkSession
from pyspark.sql.utils import AnalysisException
import pyspark.sql.functions as F
import pyspark.sql.types as T

spark = (
    SparkSession
    .builder
    .config("spark.driver.memory", "22g")
    .config("spark.executor.memory", "22g")
    .getOrCreate()
)
```

```python
train = spark.read.csv(r"C:\BigData\avazu-ctr-prediction\train\train.csv",
                      header=True,
                      inferSchema=True)
train.printSchema()
```

```python
train.show(5, truncate=True, vertical=True)
```

```python
train.count()
```

```python
num_dtypes = {"decimal(20,0)", "int"}
for col, dtype in train.dtypes:
    if dtype in num_dtypes:
        print(col)
        train.select(col).summary().show()
```

```python
str_cols = {(col, dtype) for col, dtype in train.dtypes if dtype not in num_dtypes}
str_cols
```
